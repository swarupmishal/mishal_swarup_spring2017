# Question 5
- Use Gutenberg and Web_text data. Find out what are the top 5 words that Shakespeare used but we do not use in currently.
- Take top 50 words from Shakespeare (all 3 books) and top 50 from Web_text (all the records).
- Remove punctuation and stop words.
- Remove the words we still use today, and get the unused list. Show the top 5 elements. 

```python
#import nltk
#english_vocab = set(w.lower() for w in nltk.corpus.words.words())

from nltk.corpus import webtext
webtext.fileids()
for file in webtext.fileids():
    webtext_words=webtext.words(file)
    common_words=set(w.lower() for w in webtext.words(file))
    
from nltk.corpus import stopwords
stop_words=stopwords.words('english')

caesar = list(w.lower() for w in nltk.corpus.gutenberg.words('shakespeare-caesar.txt'))
hamlet = list(w.lower() for w in nltk.corpus.gutenberg.words('shakespeare-hamlet.txt'))
macbeth = list(w.lower() for w in nltk.corpus.gutenberg.words('shakespeare-macbeth.txt'))

shakespeare_words = []
for word in caesar:
    if word.lower() not in common_words:
        if word.lower() not in stop_words:
            shakespeare_words.append(word)
        
for word in hamlet:
    if word.lower() not in common_words:
        if word.lower() not in stop_words:
            shakespeare_words.append(word)
        
for word in macbeth:
    if word.lower() not in common_words:
        if word.lower() not in stop_words:
            shakespeare_words.append(word)

#print(shakespeare_words)

uniques = []
for word in shakespeare_words:
    if word not in uniques:
        uniques.append(word)
        
counts = []
for unique in uniques:
  count = 0              # Initialize the count to zero.
  for word in shakespeare_words:     # Iterate over the words.
    if word == unique:   # Is this word equal to the current unique?
      count += 1         # If so, increment the count
  counts.append((count, unique))

counts.sort()            # Sorting the list puts the lowest counts first.
counts.reverse()         # Reverse it, putting the highest counts first.
# Print the ten words with the highest counts.
for i in range(min(5, len(counts))):
  count, word = counts[i]
  print('%s %d' % (word, count))
```

```python
#getting all the characters in all 3 Shakespeare files

total_words_sh = {}
for f in sh_list:
    c=1
    w = nltk.corpus.gutenberg.words(f)
    #printing total no of words in each file
    print('total words in', c,  'file=', len(w))
    c += 1
    #concatenating words from all 3 files
    for word in w:
        if word.isalpha():
            if word in total_words_sh:
                total_words_sh[word] +=1
            else:
                total_words_sh[word] = 1    
```

```python
#getting all the characters in all webtext file

total_words_wb = {}
for f in corpus_list_2:
    y=nltk.corpus.webtext.words()
    print('total words in file=', len(y))
    for word in y:
        if word.isalpha():
            if word in total_words_wb:
                total_words_wb[word] +=1
            else:
                total_words_wb[word] = 1 
```

```python
# sorting shakespeares words based on mostly used
import operator
total_words_sh = sorted(total_words_sh.items(),key=operator.itemgetter(1),reverse=True)
total_words_sh
```

```python
# sorting webtexts words based on mostly used
import operator
total_words_wb = sorted(total_words_wb.items(),key=operator.itemgetter(1),reverse=True)
total_words_wb
```

```python
#top 50 words used by Shakespeare collectively
top_50_sh = total_words_sh[0:49]

print('Top 50 words of Shakespeare are: ')
for word in top_50_sh:
    print(word[0])
    
#all the words used by Shakespeare
total_sh = []
for word in total_words_sh:
    total_sh.append(word[0])
```
